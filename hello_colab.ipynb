# Hello Colab — live GitHub sync test

**Pattern:** Cell 1 runs once per session. Cell 2 syncs latest code from GitHub any time. Cell 3+ re-run freely without losing runtime state.

Open via: `colab.research.google.com/github/kamalgs/specdecode/blob/master/hello_colab.ipynb`

### Cell 1 — Run once: environment check

```python
import sys, torch

print(f'Python:  {sys.version.split()[0]}')
print(f'PyTorch: {torch.__version__}')
if torch.cuda.is_available():
    print(f'GPU:     {torch.cuda.get_device_name(0)}')
    print(f'VRAM:    {torch.cuda.get_device_properties(0).total_memory / 1e9:.0f} GB')
else:
    print('GPU:     none (CPU only)')
```

### Cell 2 — Sync from GitHub (re-run any time, runtime state preserved)

```python
import os, subprocess, sys

REPO = 'https://github.com/kamalgs/specdecode.git'
REPO_DIR = 'specdecode'

if os.path.exists(REPO_DIR):
    result = subprocess.run(['git', '-C', REPO_DIR, 'pull'], capture_output=True, text=True)
    print(result.stdout.strip() or 'Already up to date.')
else:
    subprocess.run(['git', 'clone', REPO, REPO_DIR])
    print('Cloned.')

if REPO_DIR not in sys.path:
    sys.path.insert(0, REPO_DIR)

commit = subprocess.run(
    ['git', '-C', REPO_DIR, 'log', '--oneline', '-1'],
    capture_output=True, text=True
).stdout.strip()
print(f'At commit: {commit}')
```

### Cell 3 — Run code from repo (re-run after syncing)

```python
# Proves we're running the latest pushed version
with open(f'{REPO_DIR}/CONTEXT.md') as f:
    first_lines = f.readlines()[:5]

print('--- First 5 lines of CONTEXT.md (from repo) ---')
print(''.join(first_lines))
```
